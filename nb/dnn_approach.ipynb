```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
sys.path.append(os.path.abspath('../sc'))
import model_utils
import preprocess

df = pd.read_csv("../d/train.csv")
df_dev = pd.read_csv("../d/dev.csv")
df = preprocess.preprocess_texts(df)
df_dev = preprocess.preprocess_texts(df_dev)
df.head()
```

```python
tk = preprocess.fit_tokenizer(df)
sq1, sq2 = preprocess.transform_tokenizer(df, tk)
sq1_dev, sq2_dev = preprocess.transform_tokenizer(df_dev, tk)
y = df.is_duplicate.values
y_dev = df_dev.is_duplicate.values
```

```python
#filepath=os.path.join(model_utils.GLOVE_DIR, "glove.6B.100d.txt")
#weights, num_words = model_utils.load_weights(filepath, tk, embedding_dim=100)
filepath=os.path.join(preprocess.GLOVE_DIR, "glove.42B.300d.txt")
#filepath=os.path.join(model_utils.GLOVE_DIR, "glove.840B.300d.txt")
weights, num_words = preprocess.load_weights(filepath, tk)
```

```python
model_path=os.path.join("test_combined.hdf5")
model_utils.train([sq1, sq2, y], [sq1_dev, sq2_dev, y_dev], model_path, weights, num_words, epochs=30, batch_size=512)
```
